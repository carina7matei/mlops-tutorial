```python
df['duration']=df['tpep_dropoff_datetime']-df['tpep_pickup_datetime']
```

```python
df['duration']
```

```python
df['duration_minutes'] = (df['duration'].dt.total_seconds() / 60)
```

```python
df['duration_minutes'].std(ddof=1)
```

```python
df.describe()
```

```python
orig_count = len(df)
df_filtered = df[(df['duration_minutes'] >= 1) & (df['duration_minutes'] <= 60)]
```

```python
filtered_count = len(df_filtered)
```

```python
fraction_remaining = filtered_count / orig_count
fraction_remaining*100
```

```python
final_df=df_filtered[['PULocationID','DOLocationID']].copy()
```

```python
final_df['PULocationID'] = final_df['PULocationID'].astype(str)
final_df['DOLocationID'] = final_df['DOLocationID'].astype(str)
```

```python
final_df.shape
```

```python
dicts = final_df.to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=True)
X = dv.fit_transform(dicts)
```

```python
X.shape
```

```python
y = df_filtered['duration_minutes'].values
```

```python
from sklearn.linear_model  import LinearRegression

model= LinearRegression()

model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y, y_pred))
rmse
```
